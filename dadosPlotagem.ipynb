```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Nome': ['Alice', 'Jo√£o', 'Joana', 'Ana', 'Gustavo'],
        'Idade':[25, 30, 45, 23, 50],
        'Salario': [5000, 2000, 7000, 11000, 8000]}

```

```python
df =  pd.DataFrame(data)
```

```python
#plotar um grafico da barrar para sal√°rios
plt.bar(df['Nome'], df['Salario'], color='blue')
plt.title('Gr√°fico de Barras - Sal√°rios')

plt.xlabel('Nome das pessoas')
plt.ylabel('Sal√°rio (R$)')

plt.show()
```

```python
#GRAFICO DE PIZZA

plt.pie(df['Salario'], labels=df['Nome'], autopct='%1.1f%%', colors=['gold','lightcoral', 'lightskyblue', 'lightgreen', 'lightpink'])
plt.title('Distribui√ß√£o de sal√°rios')
plt.show()
                                                                   
```

```python
#grafico de dispers√£o 

plt.scatter(df['Idade'], df['Salario'], color='purple')
plt.title('Gr√°fico de Dispers√£o - Idade vs. Sal√°rio')
plt.xlabel('Idade')
plt.ylabel('Sal√°rio (R$)')
plt.show()
```

```python
#grafico de linhas para idade

plt.plot(df['Nome'], df['Idade'], marker='o', linestyle='-', color='green')
plt.title('Gr√°fico de linhas - Idade')
plt.xlabel('Nome')
plt.ylabel('Idade')
plt.show()
```

```python
#Grafico de linha para sal√°rio
plt.plot(df['Nome'], df['Salario'], marker='o', linestyle='-', color='green')
plt.title('Gr√°fico de linhas - Sal√°rio')
plt.xlabel('Nome')
plt.ylabel('Salario')
plt.show()
```
