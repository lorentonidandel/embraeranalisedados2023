```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {'Nome': ['Alice', 'João', 'Joana', 'Ana', 'Gustavo'],
        'Idade':[25, 30, 45, 23, 50],
        'Salario': [5000, 2000, 7000, 11000, 8000]}

```

```python
df =  pd.DataFrame(data)
```

```python
#plotar um grafico da barrar para salários
plt.bar(df['Nome'], df['Salario'], color='blue')
plt.title('Gráfico de Barras - Salários')

plt.xlabel('Nome das pessoas')
plt.ylabel('Salário (R$)')

plt.show()
```

```python
#GRAFICO DE PIZZA

plt.pie(df['Salario'], labels=df['Nome'], autopct='%1.1f%%', colors=['gold','lightcoral', 'lightskyblue', 'lightgreen', 'lightpink'])
plt.title('Distribuição de salários')
plt.show()
                                                                   
```

```python
#grafico de dispersão 

plt.scatter(df['Idade'], df['Salario'], color='purple')
plt.title('Gráfico de Dispersão - Idade vs. Salário')
plt.xlabel('Idade')
plt.ylabel('Salário (R$)')
plt.show()
```

```python
#grafico de linhas para idade

plt.plot(df['Nome'], df['Idade'], marker='o', linestyle='-', color='green')
plt.title('Gráfico de linhas - Idade')
plt.xlabel('Nome')
plt.ylabel('Idade')
plt.show()
```

```python
#Grafico de linha para salário
plt.plot(df['Nome'], df['Salario'], marker='o', linestyle='-', color='green')
plt.title('Gráfico de linhas - Salário')
plt.xlabel('Nome')
plt.ylabel('Salario')
plt.show()
```
